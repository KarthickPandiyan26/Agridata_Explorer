```python
#importing packages

import pandas as pd #for data reading
import seaborn as sns #for ploting
import matplotlib.pyplot as plt #for ploting
from sqlalchemy import create_engine #for DB connection
import mysql.connector #for DB connection
```

```python
#data reading

df = pd.read_csv(r"c:\Users\karth\Desktop\Data Science\DS Project\Agri Data Explorer\Indian Agircultural Data.csv")
```

```python
df.fillna(0, inplace=True)

df.columns = df.columns.str.title() # To convert the column names from Capital letters to Normal Letters
```

```python
df
```

#Exploratory Data Analysis (EDA)

```python
#Top 7 RICE PRODUCTION State Data

sns.set_style('darkgrid')
plt.figure(figsize= (7,5))

top_rice = df.groupby('State Name')['Rice Production (1000 Tons)'].sum().nlargest(7)
top_rice.plot(
    kind='bar', 
    title='Top 7 Rice-Producing States', 
    ylabel='Production (Tons)', 
    xlabel='State', 
    color='blue', 
    label='Rice Production', 
    edgecolor='black'
    )

plt.tick_params(axis='x', labelrotation=45)
plt.legend()
plt.show()
```

```python
#Top 5 Wheat Producing States Data(Bar_chart)and its percentage(%)

sns.set_style('darkgrid')
plt.figure(figsize= (8,5))

top_wheat = df.groupby('State Name')['Wheat Production (1000 Tons)'].sum().nlargest(5)
top_wheat.plot(
    kind='bar', 
    title='Top 5 Wheat-Producing States', 
    xlabel='State Name', 
    ylabel='Production (Tons)', 
    color='#FE1A76', 
    legend='Wheat Production', 
    edgecolor='black'
    )

plt.tick_params(axis='x', labelrotation=45)
plt.legend()
plt.show()
```

```python
sns.set_style('darkgrid')
plt.figure(figsize= (10,5))

top_wheat = df.groupby('State Name')['Wheat Production (1000 Tons)'].sum().nlargest(5)

top_wheat.plot(
    kind='pie', 
    autopct='%1.1f%%', 
    title='Wheat Production Percentage', 
    ylabel=('')
    )

plt.show()
```

```python
#Oil seed production by top 5 states

sns.set_style('darkgrid')
plt.figure(figsize= (8,4))

top_oil = df.groupby('State Name')['Oilseeds Production (1000 Tons)'].sum().nlargest(5)

top_oil.plot(
    kind='bar', 
    title='Top 5 Oilseed production states',
    color='#15F2A5', 
    xlabel='State', 
    ylabel='Production (Tons)', 
    label='Oilseed Production', 
    edgecolor='black'
    )

plt.tick_params(axis='x', labelrotation=45)
plt.legend()
plt.show()
```

```python
#Top 7 Sunflower Production States

sns.set_style('darkgrid')

plt.figure(figsize= (8,4))

top_sunflower = df.groupby('State Name')['Sunflower Production (1000 Tons)'].sum().nlargest(7)

top_sunflower.plot(
    kind='bar', 
    title= 'Top Sunflower Production States', 
    xlabel='State Name', 
    ylabel='Production (Tons)', 
    color='#019f81', 
    label='Sunflower Production', 
    edgecolor='black'
    )

plt.tick_params(axis='x', labelrotation=45)
plt.legend()
plt.show()
```

```python
#India's Sugarcane Production From Last 50 Years

sns.set_style('darkgrid')

plt.figure(figsize= (10,5))

sugarcane_trend = df.groupby('Year')['Sugarcane Production (1000 Tons)'].sum()
sugarcane_trend.plot(
    kind='line', 
    title='India’s Sugarcane Production (50 Years)', 
    xlabel= 'Year', 
    ylabel='Production (Tons)', 
    color='Green', 
    label='Sugarcane Production'
    )

plt.fill_between(
    sugarcane_trend.index, 
    sugarcane_trend.values, 
    color='lightgreen', 
    alpha=0.4
    )

plt.legend()
plt.show()
```

```python
#Rice Production Vs Wheat Production

sns.set_style('darkgrid')
plt.figure(figsize= (14,8))

rice_trend = df.groupby('Year')['Rice Production (1000 Tons)'].sum()
wheat_trend = df.groupby('Year')['Wheat Production (1000 Tons)'].sum()

pd.concat([rice_trend, wheat_trend], axis=1).plot(
    title='Rice vs Wheat Production (50 Years)', 
    xlabel='Year', 
    ylabel='Prodcution (Tons)', 
    color=['Red', 'Blue']
    )

plt.fill_between(
    rice_trend.index, 
    rice_trend.values, 
    color='#f5634a', 
    alpha=0.4
    )
plt.fill_between(
    wheat_trend.index, 
    wheat_trend.values, 
    color='#D0FCED', 
    alpha=0.4
    )

plt.legend(loc='upper left', title='Crop type')
plt.show()
```

```python
#Rice Production By West Bengal Districts

sns.set_style('darkgrid')
plt.figure(figsize= (10,5))

wb_rice = df[df['State Name'] == 'West Bengal'].groupby('Dist Name')['Rice Production (1000 Tons)'].sum()
wb_rice.plot(
    kind='bar', 
    title='Rice Production by West Bengal Districts', 
    xlabel='District Name', 
    ylabel='Production (Tons)', 
    color='#f5092d', 
    label='Rice Production', 
    edgecolor='black'
    )

plt.tick_params(axis='x', labelrotation=45)
plt.legend()
plt.show()
```

```python
#Top 10 Wheat Production Years From Uttar Pradesh

sns.set_style('darkgrid')

plot, ax= plt.subplots(figsize=(10,5))
ax.set_facecolor('#fff5fa')

up_wheat = df[df['State Name'] == 'Uttar Pradesh'].groupby('Year')['Wheat Production (1000 Tons)'].sum().nlargest(10)
up_wheat.plot(
    kind='bar', 
    title='Top 10 Wheat Production Years - Uttar Pradesh', 
    color= "#20e0d0", 
    xlabel='Year', 
    ylabel='Production (Tons)', 
    label='Wheat Production', 
    edgecolor='black'
    )

plt.legend()
plt.show()
```

```python
#Millet Production (Last 50y)

sns.set_style('whitegrid')

plot, ax= plt.subplots(figsize=(10,5))
ax.set_facecolor('#fff5fa')
#plt.figure(figsize= (10,5))

millet_trend= df.groupby('Year')['Finger Millet Production (1000 Tons)'].sum()
millet_trend.plot(
    kind='line', 
    title='Millet Production (50 Years)', 
    xlabel='Year', 
    ylabel='Production (Tons)', 
    label='Millet Production', 
    color='#3b0080'
    )

plt.fill_between(
    millet_trend.index, 
    millet_trend.values, 
    color="#a03ae4", 
    alpha=0.5
    )

plt.legend()
plt.show()
```

```python
#Sorghum Production (Kharif and Rabi) by Region

plt.figure(figsize= (10,5))

pivot = df.pivot_table(index='State Name', values=['Kharif Sorghum Production (1000 Tons)', 'Rabi Sorghum Production (1000 Tons)'], aggfunc='sum')
sns.heatmap(
    pivot, 
    annot=True, 
    cmap='YlGnBu', 
    fmt='.0f'
    )

plt.title('Sorghum Production Heatmap by Region')
plt.xlabel('Sorghum Production')
plt.ylabel('State')

plt.tight_layout()
plt.show()
```

```python
#Top 7 States for Groundnut Production

sns.set_style('whitegrid')

plot, ax= plt.subplots(figsize=(10,5))
ax.set_facecolor("#fffadf")

groundnut= df.groupby('State Name')['Groundnut Production (1000 Tons)'].sum().nlargest(7)
groundnut.plot(
    kind='bar', 
    title='Top 7 States for Groundnut Production', 
    xlabel='State', 
    ylabel='Production (Tons)', 
    color='#D35400', 
    label='Groundnut Production', 
    edgecolor='black' 
    )

plt.tick_params(axis='x', labelrotation=45)
plt.legend()
plt.show()
```

```python
#Soybean Production by Top 5 States and Yield Efficiency 

sns.set_style('darkgrid')
fig, ax1 = plt.subplots(figsize=(10, 5))

soyabean= df.groupby('State Name')[['Soyabean Production (1000 Tons)', 'Soyabean Yield (Kg Per Ha)']].sum().nlargest(5,'Soyabean Production (1000 Tons)' )

ax1.bar(
    soyabean.index, 
    soyabean['Soyabean Production (1000 Tons)'], 
    color="#a140dd", 
    label='Production (Tons)', 
    edgecolor='black'
    )
ax1.set_ylabel('Production (1000 Tons)')
ax1.set_xlabel('State Name')
ax1.set_title('Soybean Production & Yield Efficiency by Top 5 States')

ax2 = ax1.twinx()
ax2.plot(
    soyabean.index, 
    soyabean['Soyabean Yield (Kg Per Ha)'], 
    color="#f80000", 
    marker='o', 
    label='Yield Efficiency'
    )
ax2.set_ylabel('Yield (Kg/Ha)')

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()

ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper right')

plt.tight_layout()
plt.show()
```

```python
#Oilseed Production in Major States

sns.set_style('whitegrid')

plot, ax= plt.subplots(figsize=(10,5))
ax.set_facecolor("#fffadf")

oilseed= df.groupby('State Name')['Oilseeds Production (1000 Tons)'].sum().nlargest(15)

oilseed.plot(
    kind='bar', 
    title='Oilseed Production in Major States', 
    xlabel='State Name',
    ylabel='Production (Tons)', 
    color='#5f162b', 
    label='Oilseed Production', 
    )

plt.tick_params(axis='x', labelrotation=45)
plt.legend()
plt.show()
```

```python
#Impact of Area Cultivated on Production (Rice, Wheat, Maize) 

sns.set_style('darkgrid')

plot, ax= plt.subplots(figsize=(10,6))
ax.set_facecolor("#c9f4f1")

sns.scatterplot(
    data=df, 
    x='Rice Area (1000 Ha)', 
    y='Rice Production (1000 Tons)', 
    label='Rice', 
    color='#5b3852', 
    s=50, 
    ax=ax
    )
sns.scatterplot(
    data=df, 
    x='Wheat Area (1000 Ha)', 
    y='Wheat Production (1000 Tons)', 
    label='Wheat', 
    color="#fd12c2", 
    s=50, 
    ax=ax
    )
sns.scatterplot(
    data=df, 
    x='Maize Area (1000 Ha)', 
    y='Maize Production (1000 Tons)', 
    label='Maize', 
    color="#fb0909", 
    s=50, 
    ax=ax
    )

ax.set_xlabel('Area Cultivated (1000 Ha)')
ax.set_ylabel('Crop Production (1000 Tons)')

ax.set_title('Impact of Area Cultivated on Crop Production')
ax.legend(loc='upper right', title='Crop Type')

plt.tight_layout()
plt.show()
```

```python
#Rice vs. Wheat Yield Across States

sns.set_style('darkgrid')
fig, ax = plt.subplots(figsize=(10,6))

yield_comparison = df.groupby('State Name')[['Rice Yield (Kg Per Ha)', 'Wheat Yield (Kg Per Ha)']].mean()

yield_comparison.plot(
    kind='bar', 
    ax=ax, 
    color=['#2ca02c', '#d62728'], 
    width=0.75, 
    edgecolor='black'
    )

ax.set_title('Rice vs Wheat Yield Across States')
ax.set_xlabel('State Name', fontsize=14)
ax.set_ylabel('Yield (Kg per Ha)', fontsize=14)
ax.legend(['Rice Yield', 'Wheat Yield'], title='Crop Type')
ax.tick_params(axis='x', labelrotation=45)

plt.tight_layout()
plt.show()
```
